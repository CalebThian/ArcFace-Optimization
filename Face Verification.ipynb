```python
def verify(img1_path,img2_path):
    img1 = cv2.imread(img1_path)
    img2 = cv2.imread(img2_path)
    
    plt.imshow(img1[:,:,::-1])
    plt.show()
    plt.imshow(img2[:,:,::-1])
    plt.show()
    
    result = DeepFace.verify(img3_path,img2_path,model_name = "ArcFace")
    print("result:",result)
    verification = result['verified']
    
    if verification:
        print("They are same")
    else:
        print("They are not same")
```
